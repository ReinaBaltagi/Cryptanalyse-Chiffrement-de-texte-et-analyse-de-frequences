```python
#Question 1: Chiffrement par permutation clé nom puis prénom collés"
mot_secret="baltagireina"
#Pour parcourir chaque lettre du mot secret je vais faire une boucle "for"
#Mais avant je dois créer une liste vide pour stocker les lettres non doubles
lettres_uniques=[]
for lettre in mot_secret:
    if lettre not in lettres_uniques:
        lettres_uniques.append(lettre) #Append ajoute un élément à la fin d'une liste
#Puis je convertie la liste en chaîne de caractères avec une boucle for qui sert à ajouter chaque lettre unique du mot secret intial au mot secret final
mot_secret_final=""
for lettre in lettres_uniques:
    mot_secret_final+=lettre

print(mot_secret_final)
    
```

```python
#Le but maintenant c'est d'essayer de mettre tous les autres lettres à savoir celles qui ne figurent pas dans le mot secret final
```

```python
#je commence par définir la taille du tableau 
n_lignes= 3
n_colonnes=9

#Puis je définis l'alphabet complète en minuscules
alphabet="abcdefghijklmnopqrstuvwxyz"

#Par la suite je dois chercher les lettres à ajouter à savoir l'alphabet moins les lettres du mot secret
lettres_a_ajouter=[lettre for lettre in alphabet if lettre not in mot_secret_final]

#Puis je cherche à trouver les lettres totales à mettre dans le tableau
lettres_totales=list(mot_secret_final)+lettres_a_ajouter

#Je créé le tableau vide, qui servira comme une liste des listes
tableau=[]

#Et pour parcourir les lettres_totales je créé un index à 0 
index=0

#Finalement, je met la boucle qui me permettra de remplir le tableau vide, avec mon alphabet complète chiffré, à savoir, 
#Suivant l'ordre de l'alphabet normale mais après le mot_secret_final, en déductant à chaque fois, la lettre qui y figure

for i in range(n_lignes): #Boucle créant chaque ligne du tableau et s'exécute 3 fois puisque n_lignes = 3
    ligne=[] #On créé une liste vide à chaque passage à la ligne, qui sera remplie par la boucle suivante
    
    for j in range (n_colonnes):# Boucle qui parcours les colonnes (9 colonnes) afin de les remplir et de remplir les lignes de la plus grande boucle
        #Pour faire cela  j'ajoute une condition qui vérifie si index est toujours dans la plage des lettres disponibles 
        if index < len (lettres_totales):
            ligne.append(lettres_totales[index])#si oui, j'ajoute la lettre correspondante dans ligne 
            index +=1 #Et j'incrémente pour passer à la lettre suivante
        else:
            ligne.append(" ") #Cet espace permet de remplir les cases vides à la fin
    tableau.append(ligne) #Une fois les 9 colonne remplies j'ajoute cette ligne complète au tableau 

#Puis j'affiche le tableau complet 
for ligne in tableau: 
    print(ligne)
```

```python
#Maintenant je dois créer mon nouvel alphabet, qui suit l'ordre des colonnes, à savoir, colonne, par colonne je prend les trois lignes de cette colonne
#donc les trois lettres, puis j'ajoute ceux de la colonne suivante et ainsi de suite, pour finir par avoir une alphabet de 26 lettres 

#Je commence par définir la nouvelle variable vide qui sera remplie par la suite avec une boucle 
nouvel_alphabet = ""
#La boucle parcourt les colonnes et les lignes pour prendre la lettre à la position (i,j) et l'ajouter au nouvel alphabet 

for j in range(n_colonnes):  # je parcours les colonnes
    for i in range(n_lignes):  # je parcours les lignes
        lettre = tableau[i][j]  # Je prend la lettre à la position (i, j)
        if lettre != " ":  #  j'ignore les cases vides
            nouvel_alphabet += lettre #Je concatène, à savoir je prends l'ancienne valeur de nouvel_alphabet et je rajoute une nouvelle lettre à chaque tour de la boucle

print("Nouvel alphabet :", nouvel_alphabet)
```

```python
#Question 3:  Faire l’analyse des fréquences et déchiffrer sans utiliser la permutation secrète.

from collections import Counter #pour compter et calculer les fréquences

#Je commence par définir les lettres et les bigrammes d'ordre décroissant de leur fréquence d'apparition dans la langue française
freq_fr = "esantir"  # ordre des lettres les + fréquentes en français
bigrammes_fr = ["es", "de", "le", "re", "en", "on", "nt", "ou", "an", "er"]

#Puis je calcule les fréquences dans ce texte chiffré

lettres_uniques = [c for c in texte_chiffre if c.isalpha()]
frequences = Counter(lettres_uniques)

# je tries les lettres par fréquence décroissante
lettres_tries = [lettre for lettre, _ in frequences.most_common()]
total_lettres = sum(frequences.values())

# Je trie et affiche les lettres avec pourcentages
print("Lettres les plus fréquentes dans le texte :")
print("Lettre | Occurrences | Pourcentage")
print("-----------------------------------")

for lettre, nb in frequences.most_common(10):  # top 10
    pourcentage = (nb / total_lettres) * 100
    print(f"  {lettre}      {nb:>5}        {pourcentage:5.2f}%")

print(f"\nTotal de lettres comptées : {total_lettres}")
```

```python
#Je vais commencer par voir l'hypothèse de substitution 
#à savoir si je suit exactement l'ordre des fréquences de manière linéaire entre le texte chiffré et les fréquences françaises : (mais c'est juste une hypothèse)
substitution_hypothèse = {}

for i, lettre in enumerate(lettres_tries):
    if i < len(freq_fr):
        substitution_hypothèse[lettre] = freq_fr[i]

print("Table de substitution hypothétique (lettre_chiffrée → lettre_française) :")
print(substitution_hypothèse)
print()

```

```python
# Après avoir fait une analyse première des fréquences de lettres (toujours pas finalisée),
#je vais faire une analyse de bigrammes avant de commencer d'identifier les lettres 

#Analyse des bigrammes

def extraire_bigrammes(t):
    bigs = []
    for i in range(len(t) - 1):
        if t[i].isalpha() and t[i+1].isalpha():
            bigs.append(t[i] + t[i+1])
    return bigs

bigrammes_texte = Counter(extraire_bigrammes(texte_chiffre))
bigrammes_tries = [b for b, _ in bigrammes_texte.most_common(10)]

print("Bigrammes les plus fréquents dans le texte chiffré :")
print(bigrammes_tries)
print()

# Calcul du total pourcentage
total_bigrammes = sum(bigrammes_texte.values())

# Affichage côté à côte avec pourcentage
print("Rang | Bigramme texte | Fréquence | Bigramme français")
print("-----------------------------------------------------")

for i in range(len(bigrammes_tries)):
    bg = bigrammes_tries[i]
    freq = (bigrammes_texte[bg] / total_bigrammes) * 100
    bg_fr = bigrammes_fr[i] if i < len(bigrammes_fr) else ""
    print(f"{i+1:>4} | {bg:^15} | {freq:6.2f}%   | {bg_fr}")
```
